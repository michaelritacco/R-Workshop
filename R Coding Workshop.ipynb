## Inferential Statistics




### z-test

The z-test is used to determine whether there is a significant difference between the mean of a sample and a known population mean.

- **Null Hypothesis**: This hypothesis states that the sample mean ($\mu$) is equal to the population mean ($\mu_0$).
  
  $$H_0: \mu = \mu_0$$


- **Alternative Hypothesis**: This hypothesis states that the sample mean ($\mu$) is different from the population mean ($\mu_0$).
  
  $$H_1: \mu \neq \mu_0$$

- **Test Statistic**: The test statistic for a z-test is calculated as:

  $$Z = \frac{\bar{X} - \mu_0}{\sigma/\sqrt{n}}$$

  where $\bar{X}$ is the sample mean, $\mu_0$ is the population mean under the null hypothesis, $\sigma$ is the population standard deviation, and $n$ is the sample size.

- **Distribution Under the Null Hypothesis**: Under the null hypothesis, the test statistic $Z$ follows a standard normal distribution ($Z$-distribution):

  $$ Z \sim N(0, 1) $$


```r
# Conducting a z-test in R
# Assuming a known population mean (e.g., 180000) and using sample data

mu <- 180000  # Population mean (μ)
x_bar <- mean(housing_data$SalePrice)  # Sample mean
sigma <- sd(housing_data$SalePrice)  # Sample standard deviation
n <- length(housing_data$SalePrice)  # Sample size

# Calculating the z-score
z_score <- (x_bar - mu) / (sigma / sqrt(n))
z_score
```

### t-test




```r
# Conducting a t-test in R
# Comparing the sale prices of two different house styles
group1 <- housing_data$SalePrice[housing_data$HouseStyle == "1Story"]
group2 <- housing_data$SalePrice[housing_data$HouseStyle == "2Story"]
t_test_result <- t.test(group1, group2)
t_test_result
```

## Linear Regression

```r
# Plotting SalePrice against LotArea to visually inspect the relationship before fitting a linear model.

```

```r
# Simple Linear Regression
model <- lm(SalePrice ~ LotArea, data = housing_data)  # Linear regression model predicting Price from Area
summary(model)  # Summary of the regression model
```

```r
# Plotting the regression line on the scatter plot
plot(housing_data$LotArea, housing_data$SalePrice, main="Regression Line: SalePrice vs LotArea", xlab="Lot Area", ylab="Sale Price")
abline(model, col="red")
```

 The model's equation can be displayed using LaTeX notation:

 SalePrice = $\beta_0 + \beta_1 \cdot LotArea $
 where $\beta_0$ (Intercept) and $\beta_1$ (Slope) are estimated from the data.
